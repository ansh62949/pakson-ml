```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.preprocessing import StandardScaler

# Load MSR Train Data
msr_train = pd.read_csv('/content/train_data.csv')
msr_train.columns = msr_train.columns.str.strip().str.replace('\n', '').str.replace('\r', '')

# Identify feature columns
exclude_cols = ['subject#', 'UPDRS']
msr_columns = [col for col in msr_train.columns if col not in exclude_cols and msr_train[col].dtype != 'object']

# Discretize UPDRS into quantile-based classes
try:
    bins = pd.qcut(msr_train['UPDRS'], q=3, duplicates='drop')
    num_bins = bins.unique().size
    msr_train['UPDRS_class'] = pd.qcut(msr_train['UPDRS'], q=num_bins, labels=list(range(num_bins)))
except ValueError as e:
    print("‚ö†Ô∏è Binning failed:", e)
    msr_train['UPDRS_class'] = 0  # fallback to dummy class

# Scale features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(msr_train[msr_columns])

# Determine max LDA components
n_classes = msr_train['UPDRS_class'].nunique()
max_components = min(len(msr_columns), n_classes - 1)
print(f"‚úÖ Using {max_components} LDA component(s)")

# Apply LDA
lda = LDA(n_components=max_components)
msr_train_lda = lda.fit_transform(scaled_features, msr_train['UPDRS_class'])

# Explained variance
print('MSR train LDA: Explained variance ratio:')
print(lda.explained_variance_ratio_)

# Plot
cmap = sns.cubehelix_palette(rot=-.4, as_cmap=True)
f, ax = plt.subplots(figsize=(8, 6))
points = ax.scatter(msr_train_lda[:, 0], msr_train_lda[:, 1] if max_components > 1 else [0]*len(msr_train_lda),
                    c=msr_train['UPDRS'], s=10, cmap=cmap)
f.colorbar(points, label='UPDRS Score')
ax.set_title('LDA of MSR Train Data Colored by UPDRS')
ax.set_xlabel('LD1')
if max_components > 1:
    ax.set_ylabel('LD2')
plt.tight_layout()
plt.show()
```

```python
# Imports
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load datasets
msr_train = pd.read_csv('/content/train_data.csv')
msr_test = pd.read_csv('/content/test_data.csv')

# Clean column names
for df in [msr_train, msr_test]:
    df.columns = df.columns.str.strip().str.replace('\n', '').str.replace('\r', '')

# Identify feature columns (exclude non-numeric or target columns)
exclude_cols = ['subject#', 'UPDRS', 'class_info']
msr_columns = [col for col in msr_train.columns if col not in exclude_cols and msr_train[col].dtype != 'object']

# Scale features
scaler = StandardScaler()
scaled_train = scaler.fit_transform(msr_train[msr_columns])
scaled_test = scaler.transform(msr_test[msr_columns])

# Apply PCA
pca = PCA(n_components=10)  # You can tune this
transformed_msr_train_data = pca.fit_transform(scaled_train)
transformed_msr_test_data_forTesting = pca.transform(scaled_test)

# Train and evaluate SVM classifiers
for kernel in ['linear', 'rbf', 'poly']:
    clf = svm.SVC(kernel=kernel)
    clf.fit(transformed_msr_train_data, msr_train['class_info'])
    y_pred = clf.predict(transformed_msr_test_data_forTesting)
    acc = accuracy_score(msr_test['class_info'], y_pred)
    print(f'{kernel.capitalize()} SVM Accuracy: {acc:.4f}')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
clf.fit(transformed_msr_train_data, msr_train['class_info'])

y_pred = clf.predict(transformed_msr_test_data_forTesting)

# Correct accuracy calculation
acc = accuracy_score(msr_test['class_info'], y_pred)
print('Random Forest Accuracy: %f' % acc)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=3)
clf.fit(transformed_msr_train_data, msr_train['class_info'])
y_pred = clf.predict(transformed_msr_test_data_forTesting)
print('Decision Tree Accuracy: %f', accuracy_score(np.ones(y_pred.shape), y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(transformed_msr_train_data, msr_train['class_info'])
y_pred = gnb.predict(transformed_msr_test_data_forTesting)
print('Naive Bayes Accuracy: %f', accuracy_score(np.ones(y_pred.shape), y_pred))
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,
                    hidden_layer_sizes=(5,2), random_state=1)
clf.fit(transformed_msr_train_data, msr_train['class_info'])
y_pred = clf.predict(transformed_msr_test_data_forTesting)
print('Neural Network Accuracy: %f', accuracy_score(np.ones(y_pred.shape), y_pred))
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load combined MSR-TE training data
msrtrain_TE = pd.read_csv('/content/TE_MSRtrain_combined.csv')

# Select feature columns (columns 4 to 16 inclusive)
msr_TE_columns = msrtrain_TE.columns[4:17]

# Scale training features
scaler = StandardScaler()
scaled_msrtrain_TE = pd.DataFrame(scaler.fit_transform(msrtrain_TE[msr_TE_columns]), columns=msr_TE_columns)

# Fit PCA on scaled training data
pca = PCA()
pca.fit(scaled_msrtrain_TE)

# Function to determine intrinsic dimensionality
def intrinsic_dimension(explained_variance_ratio_, recovered_variance=0.99):
    dim = 0
    partial_var = 0
    for val in explained_variance_ratio_:
        dim += 1
        partial_var += val
        if partial_var >= recovered_variance:
            break
    return dim

# Determine number of components to retain 99% variance
n_components_msrTE = intrinsic_dimension(pca.explained_variance_ratio_)
print(f"Components making 99% of variance: {n_components_msrTE}")

# Refit PCA with optimal components
pca = PCA(n_components=n_components_msrTE)
transformed_msrtrain_TE = pca.fit_transform(scaled_msrtrain_TE)

# Load MSR test data
msr_test_data = pd.read_csv('/content/test_data.csv')

# Select overlapping features from test data
msr_test_data_TEmsr = msr_test_data[['jitter_local', 'jitter_local_absolute',
                                     'jitter_rap', 'jitter_ppq5', 'jitter_ddp', 'shimmer_local',
                                     'shimmer_local_db', 'shimmer_apq3', 'shimmer_apq5',
                                     'shimmer_apq11', 'shimmer_data', 'NTH', 'HTN']]

# Rename columns to match training set
msr_test_data_TEmsr.columns = msr_TE_columns

# Scale test features using training scaler
scaled_msr_test_data_TEmsr = pd.DataFrame(scaler.transform(msr_test_data_TEmsr), columns=msr_TE_columns)

# Transform test data using training PCA
transformed_msr_test_data_TEmsr_forTesting = pca.transform(scaled_msr_test_data_TEmsr)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load training and test data
msrtrain_TE = pd.read_csv('/content/TE_MSRtrain_combined.csv')
msr_test_data = pd.read_csv('/content/test_data.csv')

# Select feature columns from training data
msr_TE_columns = msrtrain_TE.columns[4:17]

# Scale training features
scaler = StandardScaler()
scaled_msrtrain_TE = pd.DataFrame(scaler.fit_transform(msrtrain_TE[msr_TE_columns]), columns=msr_TE_columns)

# Fit PCA on training data
pca = PCA()
pca.fit(scaled_msrtrain_TE)

# Function to compute intrinsic dimensionality
def intrinsic_dimension(explained_variance_ratio_, recovered_variance=0.99):
    dim = 0
    partial_var = 0
    for val in explained_variance_ratio_:
        dim += 1
        partial_var += val
        if partial_var >= recovered_variance:
            break
    return dim

# Determine optimal number of components
n_components_msrTE = intrinsic_dimension(pca.explained_variance_ratio_)
print(f"Components making 99% of variance: {n_components_msrTE}")

# Refit PCA with optimal components
pca = PCA(n_components=n_components_msrTE)
transformed_msrtrain_TE = pca.fit_transform(scaled_msrtrain_TE)

# Prepare test data: select matching features and rename
msr_test_data_TEmsr = msr_test_data[['jitter_local', 'jitter_local_absolute',
                                     'jitter_rap', 'jitter_ppq5', 'jitter_ddp', 'shimmer_local',
                                     'shimmer_local_db', 'shimmer_apq3', 'shimmer_apq5',
                                     'shimmer_apq11', 'shimmer_data', 'NTH', 'HTN']]
msr_test_data_TEmsr.columns = msr_TE_columns

# Scale and transform test data
scaled_msr_test_data_TEmsr = pd.DataFrame(scaler.transform(msr_test_data_TEmsr), columns=msr_TE_columns)
transformed_msr_test_data_TEmsr_forTesting = pca.transform(scaled_msr_test_data_TEmsr)

# Define training and test targets
train = transformed_msrtrain_TE
trainy = msrtrain_TE['UPDRS']  # You can switch to 'Class' if needed
test = transformed_msr_test_data_TEmsr_forTesting
testy = msr_test_data['class_info']  # This is the correct target column in test data

# Train Random Forest
rf = RandomForestRegressor(n_estimators=100, random_state=3)
rf.fit(train, trainy)
predictions = rf.predict(test)

# Evaluate
mae = mean_absolute_error(testy, predictions)
rmse = np.sqrt(mean_squared_error(testy, predictions))
print('Mean Absolute Error:', round(mae, 2))
print('Root Mean Squared Error:', round(rmse, 2))

# Plot predictions vs actual
plt.figure(figsize=(8, 5))
plt.plot(testy, predictions, '.', color='b')
plt.xlabel('Observed class_info')
plt.ylabel('Predicted UPDRS')
plt.title('Random Forest Regression: UPDRS vs class_info')
plt.grid(True)
plt.show()
```

```python
import parselmouth
from parselmouth.praat import call

sound = parselmouth.Sound("/content/tstClip.mp3")  # must be a WAV file


f0min = 75
f0max = 600
unit = "Hertz"

# Pitch and frequency features
pitch = call(sound, "To Pitch", 0.0, f0min, f0max)
meanF0 = call(pitch, "Get mean", 0, 0, unit)
stdevF0 = call(pitch, "Get standard deviation", 0, 0, unit)

# Harmonicity
harmonicity = call(sound, "To Harmonicity (cc)", 0.01, f0min, 0.1, 1.0)
hnr = call(harmonicity, "Get mean", 0, 0)

# Point process for jitter and shimmer
pointProcess = call(sound, "To PointProcess (periodic, cc)", f0min, f0max)

# Jitter features
localJitter = call(pointProcess, "Get jitter (local)", 0, 0, 0.0001, 0.02, 1.3)
localabsoluteJitter = call(pointProcess, "Get jitter (local, absolute)", 0, 0, 0.0001, 0.02, 1.3)
rapJitter = call(pointProcess, "Get jitter (rap)", 0, 0, 0.0001, 0.02, 1.3)
ppq5Jitter = call(pointProcess, "Get jitter (ppq5)", 0, 0, 0.0001, 0.02, 1.3)
ddpJitter = call(pointProcess, "Get jitter (ddp)", 0, 0, 0.0001, 0.02, 1.3)

# Shimmer features
localShimmer = call([sound, pointProcess], "Get shimmer (local)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
localdbShimmer = call([sound, pointProcess], "Get shimmer (local_dB)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
apq3Shimmer = call([sound, pointProcess], "Get shimmer (apq3)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
aqpq5Shimmer = call([sound, pointProcess], "Get shimmer (apq5)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
apq11Shimmer = call([sound, pointProcess], "Get shimmer (apq11)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
ddaShimmer = call([sound, pointProcess], "Get shimmer (dda)", 0, 0, 0.0001, 0.02, 1.3, 1.6)

# Print results
print("Mean F0:", meanF0)
print("Standard Deviation F0:", stdevF0)
print("HNR:", hnr)
print("Jitter (local):", localJitter)
print("Jitter (absolute):", localabsoluteJitter)
print("Jitter (rap):", rapJitter)
print("Jitter (ppq5):", ppq5Jitter)
print("Jitter (ddp):", ddpJitter)
print("Shimmer (local):", localShimmer)
print("Shimmer (local dB):", localdbShimmer)
print("Shimmer (apq3):", apq3Shimmer)
print("Shimmer (apq5):", aqpq5Shimmer)
print("Shimmer (apq11):", apq11Shimmer)
print("Shimmer (dda):", ddaShimmer)
```
